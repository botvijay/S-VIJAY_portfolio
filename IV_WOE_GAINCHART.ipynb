```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(model_data_selected, target_variable, test_size=0.25,random_state=23)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_model=RandomForestClassifier()
RF_model.fit(x_train,y_train)
```

```python
model_data_selected['target']=target_variable
```

```python

#binary classification models

# to evaluate performance of the model, we use gain charts or lift charts

# previous evaluation metrics: ROC curve, confusion matrix, Accuracy, Precision, Recall, f1 Score( also beta f1 score)
    
# Gain charts are used for 
1.Market Analytics: 
2.Risk Analytics
3.Supply chain management
```

```python
def model_evaluation(input_feature,model,model_data):
    predProbs = np.matrix (model.predict_proba(input_feature))
    input_feature = input_feature.assign(prob_0 = predProbs[:,0], prob_1 = predProbs[:,1]).copy()
    input_feature_1 = input_feature.assign(target = list(map(lambda x: model_data['target'][input_feature.iloc[x].name] , range (len(input_feature))))).copy()
    input_feature_2 =input_feature_1.sort_values(by = 'prob_1',ascending = False).copy()
    num_bin =10
    num_obv =len(input_feature_2)
    m = np.ceil(num_obv/num_bin).astype(np.int64)
    decile = []
    for i in range(num_bin -1):
        decile.extend([i]*m)
    decile.extend([num_bin -1]* (len(input_feature_2)-len(decile)))
    input_feature_3 = input_feature_2.assign(decile = decile).copy()
    Decile_Response = pd.crosstab(index = input_feature_3['decile'], columns = input_feature_3['target'])
    return Decile_Response 
```
